```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('apple1.jpeg')
kernel = np.ones((5,5),np.float32)/25
dst = cv2.filter2D(img,-1,kernel)
plt.figure(figsize=(20, 15))
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(dst),plt.title('Averaging')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
img = cv2.imread('apple1.jpeg')
blur = cv2.blur(img,(5,5))
plt.figure(figsize=(20, 15))

plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title('Blurred')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
img = cv2.imread('apple1.jpeg')
gau_blur = cv2.GaussianBlur(img,(5,5),0)
blur = cv2.blur(img,(5,5))
plt.figure(figsize=(20, 15))

plt.subplot(121),plt.imshow(gau_blur),plt.title('gau_blur')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title('Blurred')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
img = cv2.imread('apple1.jpeg')
median = cv2.medianBlur(img,5)
blur = cv2.blur(img,(5,5))
plt.figure(figsize=(20, 15))

plt.subplot(121),plt.imshow(median),plt.title('median')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title('Blurred')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
img = cv2.imread('apple1.jpeg')
bilater_blur = cv2.bilateralFilter(img,9,75,75)
blur = cv2.blur(img,(5,5))
plt.figure(figsize=(20, 15))

plt.subplot(121),plt.imshow(bilater_blur),plt.title('bilater_blur')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title('Blurred')
plt.xticks([]), plt.yticks([])
plt.show()
```
