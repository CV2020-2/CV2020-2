```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
img1 = cv2.imread('bike1.png')
img2 = cv2.imread('bike2.png')

gray= cv2.cvtColor(img1,cv2.COLOR_BGR2GRAY)# flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
sift = cv2.SIFT_create()
kp = sift.detect(gray,None)
img1 =cv2.drawKeypoints(gray, kp, img1)

gray= cv2.cvtColor(img2,cv2.COLOR_BGR2GRAY)# flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
sift = cv2.SIFT_create()
kp = sift.detect(gray,None)
img2 =cv2.drawKeypoints(gray, kp, img2)



plt.figure(figsize = (10, 15))
plt.subplot(1, 2, 1), plt.imshow(img1)
plt.subplot(1, 2, 2), plt.imshow(img2)
```

```python
img1 = cv2.imread('table1.png',0) # queryImage
img2 = cv2.imread('table2.png',0) # trainImage

# Initiate SIFT detector
orb = cv2.ORB()
# find the keypoints and descriptors with SIFT
kp1, des1 = orb.detectAndCompute(img1,None)
kp2, des2 = orb.detectAndCompute(img2,None)

# create BFMatcher object
bf = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck=True)
# Match descriptors.
matches = bf.match(des1,des2)
# Sort them in the order of their distance.
matches = sorted(matches, key = lambda x:x.distance)
# Draw first 10 matches.

for m, n in matches:
    if m.distance < 0.75 * n.distance:
        good.append([m])


img3 = cv2.drawMatches(img1,kp1,img2,kp2,good, flags=2)
plt.imshow(img3),plt.show()
```
