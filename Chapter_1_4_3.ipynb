```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('sIMG_2164.jpg')
res = cv2.resize(img,None,fx=2, fy=2, interpolation = cv2.INTER_CUBIC)
```

```python
height, width = img.shape[:2]
res = cv2.resize(img,(2*width, 2*height), interpolation = cv2.INTER_CUBIC)
```

```python
#cv2.imshow('img', img)
cv2.imshow('res image', res)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img = cv2.imread('sIMG_2164.jpg',0)
rows,cols = img.shape
M = np.float32([[1,0,100],[0,1,50]])
dst = cv2.warpAffine(img,M,(cols,rows))
cv2.imshow('img',dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img = cv2.imread('sIMG_2164.jpg',0)
rows,cols = img.shape
M = cv2.getRotationMatrix2D((cols/2,rows/2),90,1)
dst = cv2.warpAffine(img,M,(cols,rows))
```

```python
img = cv2.imread('sIMG_2164.jpg')
rows,cols,ch = img.shape
pts1 = np.float32([[50,50],[200,50],[50,200]])
pts2 = np.float32([[10,100],[200,50],[100,250]])
M = cv2.getAffineTransform(pts1,pts2)
dst = cv2.warpAffine(img,M,(cols,rows))
plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```

```python
img = cv2.imread('sIMG_2164.jpg')
rows,cols,ch = img.shape
pts1 = np.float32([[50,50],[200,52],[28,387],[389,390]])
pts2 = np.float32([[0,0],[300,0],[0,300],[300,300]])
M = cv2.getPerspectiveTransform(pts1,pts2)
dst = cv2.warpPerspective(img,M,(300,300))
plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```

```python
img = cv2.imread('sIMG_2164.jpg',0)
tx = 300
ty = 150
rows,cols = img.shape
M = np.float32([[1,0,tx],[0,1,ty]])
dst = cv2.warpAffine(img,M,(cols,rows))
dst = cv2.rectangle(dst, (0, 0), (tx, ty), (0, 255, 0), -1)
cv2.imshow('img',dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img = cv2.imread('sIMG_2164.jpg',0)
rows,cols = img.shape
cx = 100
cy = 100
theta = 45
M = cv2.getRotationMatrix2D((cx,cy),theta,1)
dst = cv2.warpAffine(img,M,(cols,rows))
dst = cv2.circle(dst, (cx, cy), theta, (0, 255, 0), -1)
cv2.imshow('img',dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img = cv2.imread('dsu4.jpg')
rows,cols,ch = img.shape
pts1 = np.float32([[235,457],[238,269],[569,231]])
pts2 = np.float32([[235,457],[235,270],[570,270]])
M = cv2.getAffineTransform(pts1,pts2)
dst = cv2.warpAffine(img,M,(cols,rows))
plt.figure(figsize=(20, 16))
plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```

```python
img = cv2.imread('dsu5.jpg')

rows, cols, ch = img.shape

a = [545, 529] # red
b = [540, 345] # green
c = [756, 375] # blue
d = [761, 540] # black

ta = [345, 529]
tb = [345, 345]
tc = [855, 345]
td = [860, 529]

pts1 = np.float32([a, b, c, d])
pts2 = np.float32([ta, tb, tc, td])

cv2.circle(img, tuple(a), 20, (255, 0, 0), -1)
cv2.circle(img, tuple(b), 20, (0, 255, 0), -1)
cv2.circle(img, tuple(c), 20, (0, 0, 255), -1)
cv2.circle(img, tuple(d), 20, (0, 0, 0), -1)


M = cv2.getPerspectiveTransform(pts1,pts2)
dst = cv2.warpPerspective(img,M,(cols,rows))

plt.figure(figsize = (12, 15))
plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```
