```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('hw4.png')
rows, cols, ch = img.shape
img.shape
```

```python
a = [215, 1546] # red
b = [287, 478] # green
c = [1045, 486] # blue
d = [1015, 1600] # black

ta = [215, 1546] # red
tb = [215, 478] # green
tc = [1015, 478] # blue
td = [1015, 1546] # black

pts1 = np.float32([a, b, c, d])
pts2 = np.float32([ta, tb, tc, td])

cv2.circle(img, tuple(a), 20, (255, 0, 0), -1)
cv2.circle(img, tuple(b), 20, (0, 255, 0), -1)
cv2.circle(img, tuple(c), 20, (0, 0, 255), -1)
cv2.circle(img, tuple(d), 20, (0, 0, 0), -1)

M = cv2.getPerspectiveTransform(pts1,pts2)
dst = cv2.warpPerspective(img,M,(cols,rows))

plt.figure(figsize = (12, 15))
plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```
